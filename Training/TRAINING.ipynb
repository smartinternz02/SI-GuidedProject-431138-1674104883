```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
datas = pd.read_csv('car performance dataset.csv')
```

```python
type(datas)
```

```python
datas.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
datas["car name"] = l.fit_transform(datas["car name"])
datas["horsepower"] = l.fit_transform(datas["horsepower"])
datas.head(10)
```

```python
datas.info()
```

```python
datas['car name'].unique()
```

```python
x = datas.iloc[:,1:8].values
x
```

```python
y = datas.iloc[:,0:1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x.shape
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler 
sd = StandardScaler()
x_train = sd.fit_transform(x_train)
x_test = sd.fit_transform(x_test)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestRegressor
d = RandomForestRegressor (n_estimators=30,random_state = 0)
d.fit(x_train,y_train)
```

```python
r = d.predict(x_test)
r
```

```python
import pickle
pickle.dump(d,open('regression.pkl','wb'))
```

```python
import seaborn as sns
sns.pairplot(data=datas)
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test,r)
accuracy
```
